# Decision Tree 
- Node 在節點上設定特定的 Attribute值 作為分割
- Edge 節點分割後移動到分割後結果的路徑
- Root 第一個分割的節點
- leave predict outcome
- 透過 entropy and gain function(CART algorithm) to build a tree

# Random Forest 團結力量大
- decorrelates the trees(if trees has a strong feature), reduce variance
- 優化 Decision Tree 的 Performance
- <font color="red">Build Model Process:</font>
    - step1：從 training data 透過 bagging採樣 的方式，取得部份的特徵與data後，透過CART演算法Build Tree ，然後重複此過程
    - step2：最後 Ensemble(Majority Vote)，結合所有Tree，完成了 Random Forest

- <font color="red">Bagging (Bootstrap aggregating)的注意事項</font>
    - Bagging含義：降低feature的variance(因為是取部分 feature 拿來 training，所以較Decision Tree拿全部的feature來training還要精簡)
    - Bagging方式：從training "取樣本後再放回去"， 所以平均來說會有1/3的樣本採集不到，這些稱為(Out of Bound)，並拿來衡錯誤率（冰非精準度）
    
    - 所以每棵樹從training data採集的樣本數會 < 原先所有training data的樣本數
    - 每棵樹Node數目的要求：bagging 的時候，建立的每棵樹的用到的特徵值至少要 bagging 出(k/2)+1的feature或是square(k)的特徵值(但前提是特徵值要夠多)，較有價值
- <font color="red">Ensemble 階段</font>
    - 給一筆data ，讓全部已經建立好的樹去投票
    - 若是 Regression RF ，則是加總起來 / 總棵樹，就是預估的答案

# Build Tree : Algorithm
- 常用演算法：挑選的屬性 越 pure 越好，且必須 pure than 父節點，權重 * (節點裡面的值run演算法)
    - Information Gain :ID3, C4.5,C5.0
    - Gini Function：CART
    - X^2 獨立性檢定 CHAID<br><br>
- ID3:在一開始選 Gain(attribute) 最大的屬性 Attribute  會較好
    - Gain(A) = Info(D)-Info(A)
    - Info(A) 越小越好， Gain(A)越大越好
- C4.5:看 Gain Ratio，越大越適合挑那個屬性
    - Gain(A)/SplitInfo(A)
- Tree Puning ： resolve overfitting 的問題


# Machine Learning algorithm
- from sklearn.tree import DecisionTreeClassifier 
- from sklearm.ensemble import RandomForestClassifier
- label 可以不用mapping成數值的型態

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as ply
import seaborn as sns
from sklearn.cross_validation import train_test_split
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.preprocessing import StandardScaler
import plotly.plotly as py
from plotly.offline import init_notebook_mode,iplot,plot,download_plotlyjs
import cufflinks as cf
%matplotlib inline

from sklearn.tree import DecisionTreeClassifier
```

```python
df =pd.read_csv("kyphosis.csv")
df.head()
```

```python
sns.pairplot(data=df,hue="Kyphosis")
```

```python
pd.get_dummies(df["Kyphosis"])
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
X=df.drop("Kyphosis",axis=1)
y=df["Kyphosis"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
y_test
```

# Decision Tree

```python
dtree=DecisionTreeClassifier()
dtree.fit(X_train,y_train)
predict=dtree.predict(X_test)
```

```python
print(classification_report(predict,y_test))
print(confusion_matrix(predict,y_test))
```

# Random Forest
- 參數：n_estimator 決定要建立幾棵樹

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RF=RandomForestClassifier(n_estimators=200,criterion="entropy") 
RF.fit(X_train,y_train)
predict=RF.predict(X_test)
```

```python
print(confusion_matrix(predict,y_test))
print(classification_report(predict,y_test))
```
