# Pandas Built-in Data
- dataframe.plot.{graph_type} 可以直接繪製圖表
- plot(kind="")
- plot.hist
- plot.bar(stacked=T/F)
- plot.line
- plot.scatter
- plot.hexbin(跟Scatter很像，但是圖形是六角形)
- plot.area()
- plot.kde()
- plot.density()
- plot.box()

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1=pd.read_csv("df1",index_col=0)
```

```python
import seaborn as sns

df1["A"].hist(bins=30)
```

```python
df1["A"].plot(kind="hist",bins=30)
```

```python
df1["A"].plot.hist()
```

```python
df2=pd.read_csv("df2")
```

```python
df2
```

```python
# 假想他都是先從座標軸建立面積，然後再疊上去
df2.plot.area()
```

```python
# Categorical data
# 長條圖可以疊起來 stack
df2.plot.bar()
df2.plot.bar(stacked=True)
```

```python
# Frequenct distribution
df1["A"].plot.hist(bins=30)
```

```python
df1.plot.line(x=df1.index,y="B",figsize=(12,3),lw=1)
```

```python
# 三維資料，第三維資料可能用 顏色depth  顏色是欄位C 但欄位C剛好為數值，可以轉換為顏色的參數
df1.plot.scatter(x="A",y="B",c="C",figsize=(12,9))
```

```python
df2.plot.box()
```

```python
df3=pd.DataFrame(np.random.randn(1000,2),columns=["A","B"])
```

```python
df3.plot.hexbin(x="A",y="B",gridsize=26,cmap="coolwarm")
```

```python
df2["a"].plot.kde()
```
