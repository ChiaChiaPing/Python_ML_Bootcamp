# Matplotlib part 1
- build figure() => assign axes(繪圖區)

```python
import pandas as pd
import numpy as np

# 之前用的matplot 是用 pandas 內的matplot
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.linspace(0,5,11)
y=x **2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,"r")
plt.title("y=x^2")
plt.xlabel("X Label")
plt.ylabel("Y Label")
```

```python
plt.title
```

```python
#先給出子圖的空間，然後再繪圖（ plot ）
plt.subplot(1,2,1)
plt.plot(x,y,"r")

plt.subplot(1,2,2)
plt.plot(y,x,"b")
```

```python
#如果都繪圖在同一個地方的話，graph 會合併起來
#對了向這種組合的話要一次執行完一個cell，分開執行會沒辦法
plt.subplot(1,2,1)
plt.plot(x,y,"r")

plt.subplot(1,2,1)
plt.plot(y,x,"b")
```

```python
# 自己建立透過 figure() 建立繪圖區
fig = plt.figure()
#fig.add_axes([rect])
# rect[x,y,width,height]
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel("X label")
axes.set_ylabel("Y Label")
axes.set_title("Set Title")
```

```python
fig = plt.figure()

# 設0.1不設0的原因，是因為設 0 的話，他的頂點，兩條線會交叉
# fig.add_axes() : 是代表建立一個繪圖區
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x,"r")

axes2.set_title("Smaller Plot")
axes1.set_title("Bigger Plot")
```

# Matplotlib part 2

```python
#一個fig，很多個axes（ 很多個繪圖區 ），可以運用 axes.plot() 去做資料視覺化
fig,axes=plt.subplots(nrows=3,ncols=3)

#plt.tight_layout() # 讓排版看起來更簡潔

for cur_ax in axes:
    for col in cur_ax:
        col.plot(x,y)

# 可以針對局部的子繪圖區進行繪圖
axes[0][1].plot(x,y)
axes[0][1].set_title("Title")

plt.tight_layout()
```

# Figure Size and DPI

```python
#建立一個最大的繪圖區
#裡面的paramter(figuresize=(長：寬),dpi：像素)
fig=plt.figure(figsize=(8,2),dpi=100)
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
#針對每個子axes，都配置一個符合fgsize比的圖表
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(8,2))
axes[1].plot(x,y)
axes[0].plot(y,x)
plt.tight_layout()
```

```python
# 儲存圖片
fig=plt.figure()
aw=fig.add_axes([0,0,1,1])
aw.plot(x,x**2,label="X Squared")
aw.plot(x,x**3,label="X Cubed")

#legend 是秀出類別資料，裡面可以帶他的位置
aw.legend(loc=(0.1,0.1))
fig.savefig("My_picture.png",dpi=200)
```
