# Seaborn plot introduction
### 資料間的分佈為何 Distribution Plot
- distplot
- jointplot
- pairplot
- rugplot
- kdeplot

```python
import seaborn as sns
import pandas as pd
```

```python
%matplotlib inline
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
tips.head(3)
```

```python
# univeriate means one variable
# kernel density estimation : 將數據轉換成分佈圖，底下面積為1
# bins 是分組數
sns.distplot(tips["total_bill"],kde=False,bins=30)
```

```python
#可以兩個變數(binvariate)做 joint
sns.jointplot(x="total_bill",y="tip",data=tips,kind="reg")
```

```python
# 會針對整個dataframe的 "數值資料"，show這些數值資料欄位彼此關係的 biiveriate distribution
# hue 可以將資料分類，可以選適合當category的資料欄位
sns.pairplot(tips,hue="sex",palette="coolwarm",diag_kind="kde")
```

```python
sns.rugplot(tips["total_bill"])
#越密集的地方就越高
```

```python
sns.distplot(tips["total_bill"],kde=False)
```

```python
# Don't worry about understanding this code!
# It's just for the diagram below
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

#Create dataset
dataset = np.random.randn(25)

# Create another rugplot
sns.rugplot(dataset);

# Set up the x-axis for the plot
x_min = dataset.min() - 2
x_max = dataset.max() + 2

# 100 equally spaced points from x_min to x_max
x_axis = np.linspace(x_min,x_max,100)

# Set up the bandwidth, for info on this:
url = 'http://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth'

bandwidth = ((4*dataset.std()**5)/(3*len(dataset)))**.2


# Create an empty kernel list
kernel_list = []

# Plot each basis function
for data_point in dataset:
    
    # Create a kernel for each point and append to list
    kernel = stats.norm(data_point,bandwidth).pdf(x_axis)
    kernel_list.append(kernel)
    
    #Scale for plotting
    kernel = kernel / kernel.max()
    kernel = kernel * .4
    plt.plot(x_axis,kernel,color = 'grey',alpha=0.5)

plt.ylim(0,1)
```

```python
# To get the kde plot we can sum these basis functions.

# Plot the sum of the basis function
sum_of_kde = np.sum(kernel_list,axis=0)

# Plot figure
fig = plt.plot(x_axis,sum_of_kde,color='indianred')

# Add the initial rugplot
sns.rugplot(dataset,c = 'indianred')

# Get rid of y-tick marks
plt.yticks([])

# Set title
plt.suptitle("Sum of the Basis Functions")
```

```python
sns.distplot(tips["total_bill"],hist=False,rug=True,kde=True)
```

```python
sns.kdeplot(tips["total_bill"])
```

# Categorical Plot
- barplot
- countplot (只能放一個維度的資料)
- boxplot
- violin plot
- swarmplot
- factorplot(可以繪製所有的categorical data ，調整參數的kind就可)
- 參數可以搭配 hue 或是 split or color

```python
import numpy as np
tips=sns.load_dataset("tips")
tips.head(3)
```

```python
# show total_bill average
# 可以透過extimator改變 y(欲比較的數值)
# barplot 適用於 單變量資料的呈現
fig=sns.barplot(x="sex",y="total_bill",data=tips,estimator=np.mean)
fig.set_ylabel("Mean(total_bill)")
```

```python
#與barplot不一樣的是 countplot 純粹顯示出現的個數而已
# barplot 的 y軸可以是很多種資料，像是平均數，標準差等等
# countplot只能放一個維度的資料
sns.countplot(x="sex",data=tips)
```

```python
# 盒狀圖 四分位距
sns.boxplot(x="day",y="total_bill",data=tips,hue="smoker")
```

```python
# 針對 所有的 data points 做 kde
sns.violinplot(x="day",y="total_bill",data=tips,hue="sex",split=True)
```

```python
# show all the datapoint than violin and boxplot
sns.stripplot(x="day",y="total_bill",data=tips,hue="sex",split=True,jitter=True)
```

```python
# like combine the violinplot and stripplot
sns.violinplot(x="day",y="total_bill",data=tips)
sns.swarmplot(x="day",y="total_bill",data=tips,color="black")
```

```python
sns.factorplot(x="day",y="total_bill",data=tips,kind="swarm",estimator=np.mean)
```

# metrix plot
- heatmap 是用來呈現矩陣圖形的，通常會是描述兩變數間的關係
- clustermap

```python
tips=sns.load_dataset("tips")
flights=sns.load_dataset("flights")
tips.head()
```

```python
flights.head()
```

```python
#先選取資料表的數值資料的欄位，顯示彼此兩兩變數間的相關係數
tc=tips.corr()
tc
```

```python
# metric plot
# annot : annotation 註解
sns.heatmap(tc,annot=True,cmap="coolwarm")
```

```python
flights.head(3)
```

```python
fp=flights.pivot_table(index="month",columns="year",values="passengers")
```

```python
#從圖就可以知道 在 19607月的時候，搭乘的乘客比較多
fg=sns.heatmap(fp,square=True,cmap="coolwarm",linecolor="white",linewidths=0.2)
fig=fg.get_figure()
fig.savefig("123.png",dpi=300)
```

```python
# 把相似的數值所屬類別 => 階層群組化
# normalization
sns.clustermap(fp,cmap="coolwarm",standard_scale=1)
```

# Grid
- PairGrid
- FacetGrid
- 彈性的繪製圖表，不會太過單一同一圖表

```python
iris=sns.load_dataset("iris")
iris.head(5)
```

```python
sns.pairplot(iris)
```

```python
#PairGrid與 pairplot類似，但彈性比pairplot大，因為他的子圖可以自定義
g=sns.PairGrid(iris)
g.map_diag(sns.distplot) # 針對一維變量
g.map_upper(plt.scatter) #針對上半面雙變量
g.map_lower(sns.kdeplot) #針對下半面雙變量
```

```python
tips=sns.load_dataset("tips")
tips.head(5)
```

```python
# 與matplotlib 一開始繪圖的感覺很像
g = sns.FacetGrid(data=tips,col="time",row="smoker") #有點像是在分組的感覺，然後以column 和 row的組合為基底，看與其他變數的關係
g.map(sns.distplot,"total_bill")
```

```python
g = sns.FacetGrid(data=tips,col="time",row="smoker") #有點像是在分組的感覺，然後以column 和 row的組合為基底，看與其他變數的關係
g.map(plt.scatter,"total_bill","tip")
```

# Regression Plot
- lmplot
- 參數的col 與 row 代表你要的分佈是基於哪一個 row,col 群組
- 只有 lmplot 與 facetgrid 可以調整 繪圖區的 col 和 row

```python
tips=sns.load_dataset("tips")
```

```python
#透過markers 分類兩個
sns.lmplot(x="total_bill",y="tip",data=tips,hue="sex",markers=['o','v'],scatter_kws={'s':100}) #scatter_kws call matplotlib
```

```python
#呈現在同一張圖上
sns.lmplot(x="total_bill",y="tip",data=tips,hue='sex')
```

```python
#呈現在兩張圖上
#只是把資料分成col*row組，探討關係而已
sns.lmplot(x="total_bill",y="tip",data=tips,col='day',hue='sex',aspect=0.6,size=8)
```
