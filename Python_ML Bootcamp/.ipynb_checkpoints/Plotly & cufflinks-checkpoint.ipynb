# Plotly & cufflinks
- iplot(kind={bubble,spread,surface})
- df.scatter_matrix

```python
# open_source
# plotly is an interactive visuialization library
# cufflinks connect the plotly library and pandas
```

```python
import pandas as pd
import numpy as np
from plotly import __version__
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.graph_objs as go
```

```python
import cufflinks as cf
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
```

```python
# connect interactive js chart
init_notebook_mode(connected=True)
```

```python
# 之所以 pandas 可以直接用 plotly的套件，是因為 有 cufflinks
cf.go_offline()
```

```python
df1=pd.DataFrame(np.random.randn(100,4),columns="A B C D".split())
df2=pd.DataFrame({"Category":["A","B","C"],"Values":[32,43,58]})
```

```python
df1
```

```python
df1["A"].plot()
```

```python
df1["A"].iplot()
```

```python
df1.iplot(x="A",y="B",mode='markers',size=20)
```

```python
df2.iplot(x="Category",y="Values",kind="bar")
```

```python
df1.sum().iplot(kind="bar")
```

```python
df1[["A","C"]].iplot(kind="box")
```

```python
df3=pd.DataFrame({"x":[1,2,3,4,5],"y":[10,20,30,20,10],"z":[500,400,300,200,100]})
```

```python
df3.iplot(kind="surface",colorscale="rdylbu")
```

```python
fig=df1.iplot(kind="hist",bins=50)
```

```python
# 可以適用於股價的資料 (spread)，兩者折線內容的差異
df1[["A","B","C"]].iplot(kind="spread")
```

```python
df1.iplot(kind="bubble",x="A",y="B",size="C")
```

```python
df1.scatter_matrix()
```
